```python
import pandas as pd
ratings = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Movie Recommendation System/datasets/ratings.csv")
movies_metadata = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Movie Recommendation System/datasets/movies_metadata.csv")
```

```python
ratings.isnull().sum()
```

```python
movies_metadata.isnull().sum()
```

```python
movies_metadata = movies_metadata[['id','title','vote_average','vote_count','popularity']]
```

```python
movies_metadata
```

```python
movies_metadata.sort_values(by = 'vote_count',ascending=False)
```

```python
ratings.rename(columns={'movieId' : 'id'},inplace = True)
ratings['id'] = ratings['id'].astype('str')
```

```python
merged_data = ratings.merge(movies_metadata,on='id')
```

```python
merged_data.shape
```

```python
merged_data = merged_data[merged_data['vote_count'] >= 7000].sort_values('vote_average',ascending=False)
```

```python
merged_data.drop_duplicates(subset=['id'],inplace = True)
```
